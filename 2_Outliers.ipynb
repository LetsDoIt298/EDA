## Using z-score

If data is after 3rd standard deviation it can be termed as outliers

```python


def detect(data):
    outlier=[]
    thres = 3
    mean = np.mean(data)
    std = np.std(data)
    
    for i in data:
        z_score = (i-mean)/std
        if np.abs(z_score) > thres:
            outlier.append(i)
            
    return outlier
detect(dataset)
```

# IQR

```python
def outlier_iqr(data):
    outlier = []
    q1 = np.percentile(data,[25])[0]
    q3 = np.percentile(data,[75])[0]
    iqr = q3-q1
    
    lower_fence = q1 - iqr*3.5
    upper_fence = q3 + iqr*3.5
    
    for i in data:
        if i<lower_fence or i>upper_fence:
            outlier.append(i)
    return outlier

data = sorted(dataset)
outlier_iqr(data)
```
